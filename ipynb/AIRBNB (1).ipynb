```python
import pandas as pd
import numpy as np

# Cargar el CSV
df = pd.read_csv('/content/dataset_airbnb_data_lista_actualizado2.csv')

# Si la columna 'name', 'host_name' o 'room_type' están vacías (nulas), las rellenamos con 'unknown'
df['name'].fillna('unknown', inplace=True)
df['host_name'].fillna('unknown', inplace=True)
df['room_type'].fillna('unknown', inplace=True)

# Convertir la columna 'last_review' a tipo fecha
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

# Imputar las fechas faltantes con la mediana de las fechas existentes
median_date = df['last_review'].median()
df['last_review'].fillna(median_date, inplace=True)

# Corregir valores erróneos en 'reviews_per_month'
def corregir_reviews_per_month(valor):
    try:
        return float(valor)
    except ValueError:
        return np.nan

df['reviews_per_month'] = df['reviews_per_month'].apply(corregir_reviews_per_month)

# Imputar 'reviews_per_month' faltantes con la mediana
median_reviews_per_month = df['reviews_per_month'].median()
df['reviews_per_month'].fillna(median_reviews_per_month, inplace=True)

# Función para calcular la cantidad de meses entre dos fechas
def calcular_meses(fecha_inicio, fecha_final):
    if pd.isna(fecha_final):
        return np.nan
    return (fecha_final.year - fecha_inicio.year) * 12 + (fecha_final.month - fecha_inicio.month)

# Definir la fecha de inicio
fecha_inicio = pd.to_datetime('2020-01-01')

# Imputar 'number_of_reviews' cuando es nulo
def imputar_reviews(row):
    if pd.isna(row['number_of_reviews']) and pd.notna(row['last_review']) and pd.notna(row['reviews_per_month']):
        meses = calcular_meses(fecha_inicio, row['last_review'])
        if pd.notna(meses) and meses >= 0:
            return meses * row['reviews_per_month']
    return row['number_of_reviews']

# Aplicar la función de imputación para 'number_of_reviews'
df['number_of_reviews'] = df.apply(imputar_reviews, axis=1)

# Imputar los nulos restantes en 'number_of_reviews' con la mediana
median_number_of_reviews = df['number_of_reviews'].median()
df['number_of_reviews'].fillna(median_number_of_reviews, inplace=True)

# ------------------- Imputación de 'price' usando 'neighbourhood' y 'neighbourhood_group' -------------------

# Verificar que 'price' sea numérico
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Calcular la mediana del precio para cada 'neighbourhood' y 'neighbourhood_group'
median_price_by_neighbourhood = df.groupby('neighbourhood')['price'].median()
median_price_by_neighbourhood_group = df.groupby('neighbourhood_group')['price'].median()

# Lista de id problemáticos
id_intercambio_nuevos = [
    36487262, 36487263 , 36487265 , 36487267 , 36487283 , 36487289 , 36487296 , 36487309 , 36487310 , 36487311 ,
    36487312 , 36487318 , 36487319  , 36487323 , 36487329 , 36487332 , 36487341 , 36487342  , 36487343 , 36487344  ,
    36487348 , 36487349 , 36487350 , 36487353  , 36487354 , 36487355 , 36487358  , 36487359 , 36487361 , 36487362  ,
    36487363 , 36487364 , 36487366 , 36487368  , 36487369 , 36487370 , 36487372 , 36487373 , 36487376 , 36487378 ,
    36487382
]

# Función para imputar valores nulos en 'price'
def imputar_price(row):
    if pd.isna(row['price']):
        # Intentar imputar con la mediana de 'neighbourhood'
        price_imputado = median_price_by_neighbourhood.get(row['neighbourhood'], np.nan)

        # Si no se pudo imputar con 'neighbourhood', intentar con 'neighbourhood_group'
        if pd.isna(price_imputado):
            price_imputado = median_price_by_neighbourhood_group.get(row['neighbourhood_group'], np.nan)

        return price_imputado
    return row['price']

# Aplicar la función para imputar los valores nulos de 'price'
df['price'] = df.apply(imputar_price, axis=1)

# ------------------------------------------------------------------------------------

# Guardar el DataFrame modificado en un archivo Excel
df.to_excel('/content/dataset_airbnb_data_imputado_nuevos.xlsx', index=False)

# Verificar cuántos valores siguen siendo nulos
nulos_number_of_reviews = df['number_of_reviews'].isna().sum()
nulos_price = df['price'].isna().sum()

print(f"Archivo Excel guardado. Todavía quedan {nulos_number_of_reviews} valores nulos en 'number_of_reviews'.")
print(f"Y todavía quedan {nulos_price} valores nulos en 'price'.")
```

Identificación de outliers para variables númericas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV
df = pd.read_csv('/content/dataset_airbnb_data_imputado_nuevos.csv')

# Filtrar las columnas numéricas
columnas_numericas = df.select_dtypes(include=[np.number]).columns

# Definir función para identificar outliers utilizando el rango intercuartil (IQR)
def identificar_outliers(df, columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    outliers = df[(df[columna] < limite_inferior) | (df[columna] > limite_superior)]
    return outliers

# Crear un diccionario para almacenar los outliers por cada columna numérica
outliers_dict = {}

# Identificar outliers en todas las columnas numéricas
for columna in columnas_numericas:
    outliers = identificar_outliers(df, columna)
    outliers_dict[columna] = outliers

# Mostrar la cantidad y el porcentaje de outliers por cada variable numérica
for columna, outliers in outliers_dict.items():
    porcentaje_outliers = (len(outliers) / len(df)) * 100
    print(f'La variable "{columna}" tiene {len(outliers)} outliers, lo que representa el {porcentaje_outliers:.2f}% del total de datos.')

# Visualizar los outliers con gráficos de caja (boxplots)
for columna in columnas_numericas:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[columna])
    plt.title(f'Outliers en "{columna}"')
    plt.show()

```

```python
import pandas as pd
import numpy as np

# Función para identificar si un valor es un outlier utilizando el rango intercuartil (IQR)
def es_outlier(df, columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    return (df[columna] < limite_inferior) | (df[columna] > limite_superior)

# Crear dos DataFrames para datos con outliers y datos normales
df_outliers = pd.DataFrame()  # Para guardar registros con outliers
df_normales = df.copy()       # Para guardar registros sin outliers

# Identificar outliers en cada columna numérica y segmentar los datos
for columna in columnas_numericas:
    # Seleccionar registros con outliers en la columna actual
    outliers_columna = es_outlier(df, columna)

    # Alinear los índices de los DataFrames antes de aplicar el filtro booleano
    outliers_columna = outliers_columna.reindex(df.index)

    # Agregar los outliers al DataFrame de outliers
    df_outliers = pd.concat([df_outliers, df[outliers_columna]], ignore_index=True)

    # Filtrar registros sin outliers en la columna actual
    df_normales = df_normales.loc[~outliers_columna]

# Ver cuántos registros están en cada segmento sin eliminar duplicados
print(f"Registros con outliers: {len(df_outliers)}")
print(f"Registros sin outliers: {len(df_normales)}")
```

```python
import pandas as pd
from scipy.stats import zscore

# Función para detectar outliers en un DataFrame
def detectar_outliers(df):
    # Aplicar la puntuación z a las columnas numéricas
    for col in df.select_dtypes(include=['float64', 'int64']).columns:
        df[col + '_zscore'] = zscore(df[col])
        # Identificar outliers con un umbral de puntuación z > 3
        df[col + '_outliers'] = (abs(df[col + '_zscore']) > 3)
    return df

# Leer los archivos CSV
hosts_clean = pd.read_csv('/content/hosts_clean.csv')
reviews_clean = pd.read_csv('/content/reviews_clean.csv')
rooms_clean = pd.read_csv('/content/rooms_clean.csv')

# Detectar outliers en cada DataFrame
hosts_clean_outliers = detectar_outliers(hosts_clean)
reviews_clean_outliers = detectar_outliers(reviews_clean)
rooms_clean_outliers = detectar_outliers(rooms_clean)

# Verificar cuántos outliers hay en cada columna numérica
print("Outliers en hosts_clean:")
print(hosts_clean_outliers.filter(like='_outliers').sum())

print("Outliers en reviews_clean:")
print(reviews_clean_outliers.filter(like='_outliers').sum())

print("Outliers en rooms_clean:")
print(rooms_clean_outliers.filter(like='_outliers').sum())

# Función para contar duplicados en cada columna de un DataFrame
def contar_duplicados(df):
    duplicados_por_columna = df.apply(lambda x: x.duplicated().sum())
    return duplicados_por_columna

# Contar duplicados en cada DataFrame
duplicados_hosts_clean = contar_duplicados(hosts_clean)
duplicados_reviews_clean = contar_duplicados(reviews_clean)
duplicados_rooms_clean = contar_duplicados(rooms_clean)

# Mostrar el conteo de duplicados por columna en cada DataFrame
print("Duplicados en hosts_clean:")
print(duplicados_hosts_clean)
print("Duplicados en reviews_clean:")
print(duplicados_reviews_clean)
print("Duplicados en rooms_clean:")
print(duplicados_rooms_clean)


```

```python
import pandas as pd

# Cargar el archivo Excel
file_path = '/content/rooms_clean_fixed.xlsx'  # Asegúrate de que la ruta es correcta
df = pd.read_excel(file_path)

# Eliminar los puntos de las columnas latitude y longitude
df['latitude'] = df['latitude'].astype(str).str.replace('.', '', regex=False)
df['longitude'] = df['longitude'].astype(str).str.replace('.', '', regex=False)

# Convertir las columnas latitude y longitude a números, forzando los no numéricos a NaN
df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
df['longitude'] = pd.to_numeric(df['longitude'], errors='coerce')

# Corregir los valores dividiendo por 10^5
df['latitude'] = df['latitude'] / 10**5
df['longitude'] = df['longitude'] / 10**5

# Mostrar las filas donde los valores de latitude o longitude son NaN para revisarlas
nans = df[df['latitude'].isna() | df['longitude'].isna()]
print("Filas con NaN en latitude o longitude después de la corrección:")
print(nans[['latitude', 'longitude']])

# Guardar el archivo corregido en un nuevo Excel
output_excel = '/content/rooms_clean_fixed_corrected.xlsx'
df.to_excel(output_excel, index=False)

# Guardar el archivo corregido en un CSV
output_csv = '/content/rooms_clean_fixed_corrected.csv'
df.to_csv(output_csv, index=False)

```
